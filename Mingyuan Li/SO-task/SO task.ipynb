```python
def detect_stoch(df, k, d):    
    #buy_price = []
    #sell_price = []
    stoch_signal = []
    signal = 0 
    # 0 Default signal
    # f green/red cross over either in bottom 20% zone.
    # e green/red cross over either in top 20% zone.
    # a k line touch  80
    # b d line touch 80
    # c k line touch 20
    # d d line touch 20

    for i in range(0,len(df)):
        if i > 0:
       
            if k[i]>=80  and k[i-1]<80 :
                    signal = 'a'
                    stoch_signal.append(signal)
            elif k[i]>=80  and k[i-1]<80 :
                    signal = 'a'
                    stoch_signal.append(signal)

            elif d[i]>=80 and d[i-1]<80 :
                    signal = 'b'
                    stoch_signal.append(signal)

            elif k[i]<=20 and k[i-1]>20 :
                    signal = 'c'
                    stoch_signal.append(signal)

            elif d[i]<=20 and d[i-1]>20 :
                    signal = 'd'
                    stoch_signal.append(signal)

            else:

                stoch_signal.append(0)
        else:
            stoch_signal.append(0)
            
            
    return stoch_signal
```

```python
def detect_stoch_ef(df, k, d):    
    #buy_price = []
    #sell_price = []
    stoch_signal = []
    signal = 0 
    # 0 Default signal
    # f green/red cross over either in bottom 20% zone.
    # e green/red cross over either in top 20% zone.
    # a k line touch  80
    # b d line touch 80
    # c k line touch 20
    # d d line touch 20

    for i in range(0,len(df)):
        if i>0:
            if k[i] < 20 and d[i] < 20 and ((k[i] <d[i] and k[i-1] > d[i-1]) or (k[i] > d[i] and k[i-1] < d[i-1])): # green/red cross over either in bottom 20% zone.
                    signal = 'f'
                    stoch_signal.append(signal)
                    
            elif k[i] > 80 and d[i] > 80 and ((k[i] >d[i] and k[i-1] < d[i-1]) or (k[i] <d[i] and k[i-1] > d[i-1])): # green/red cross over either in top 20% zone.
                    
                    signal = 'e'
                    stoch_signal.append(signal)
            else:

                stoch_signal.append(0)
        else:
            stoch_signal.append(0)
            
    return stoch_signal
```

```python
def detectlable(filename,k,d):
    filename=filename+'.csv'
    df = pd.read_csv("/Users/limingyuan/Desktop/Data_OANDA/"+filename)
    df.columns=['datetime', 'open', 'high', 'low', 'close']
    df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%dT%H:%M:%S.000000000Z')
    
    df.ta.stoch(high='high', low='low', k=k, d=d, append=True)
    
    k=df.columns[-1]
    d=df.columns[-2]
    detect_stochef=detect_stoch_ef(df, df[k].iloc, df[d].iloc)
    detect_stochl=detect_stoch(df, df[k].iloc, df[d].iloc)
    
    df_concat=pd.concat([df, pd.DataFrame(detect_stochl,columns=['label1'])],axis=1)
    df_concat=pd.concat([df_concat, pd.DataFrame(detect_stochef,columns=['label2'])],axis=1)
    
    df_concat.to_excel("/Users/limingyuan/Desktop/Data_OANDA/SO/"+filename+".xlsx",header=True,index=False)
    
    
    
    
    
```

```python
file=['AUD_USD_M15_2005_202206','EUR_USD_M15_2005_202206','GBP_USD_M15_2005_202206','USD_CAD_M15_2005_202206','USD_JPY_M15_2005_202206',
     'USD_CHF_M15_2005_202206','NZD_USD_M15_2005_202206','AUD_CAD_M15_2005_202206']

for i in file:
    detectlable(i,14,3)
```

```python
df=df_concat[:300]

df.columns = [x.lower() for x in df.columns]
# Create our primary chart
# the rows/cols arguments tell plotly we want two figures
fig = make_subplots(rows=2, cols=1)  
# Create our Candlestick chart with an overlaid price line
fig.append_trace(
    go.Candlestick(
        x=df['datetime'],
        open=df['open'],
        high=df['high'],
        low=df['low'],
        close=df['close'],
        showlegend=False
    ), row=1, col=1  # <------------ upper chart
)
# price Line
fig.append_trace(
    go.Scatter(
        x=df['datetime'],
        y=df['open'],
        line=dict(color='#ff9900', width=1),
        name='open',
    ), row=1, col=1  # <------------ upper chart
)

# Fast Signal (%k)
fig.append_trace(
    go.Scatter(
        x=df['datetime'],
        y=df['stochk_14_3_3'],
        line=dict(color='green', width=2),
        name='%k',
    ), row=2, col=1  #  <------------ lower chart
)
# Slow signal (%d)
fig.append_trace(
    go.Scatter(
        x=df['datetime'],
        y=df['stochd_14_3_3'],
        line=dict(color='red', dash='dot',width=2),
        name='%d'
    ), row=2, col=1  #<------------ lower chart
)

ax=[]
at=[]
bx=[]
bt=[]
cx=[]
ct=[]
dx=[]
dt=[]
ex=[]
et=[]
fx=[]
ft=[]

for i in range(0,len(df)):
    if df['label1'].iloc[i]== 'a':
        ax.append(df['stochk_14_3_3'].iloc[i])
        at.append(df['datetime'].iloc[i])
    if df['label1'].iloc[i]== 'b':
        bx.append(df['stochd_14_3_3'].iloc[i])
        bt.append(df['datetime'].iloc[i])
    if df['label1'].iloc[i]== 'c':
        cx.append(df['stochk_14_3_3'].iloc[i])
        ct.append(df['datetime'].iloc[i])
    if df['label1'].iloc[i]== 'd':
        dx.append(df['stochd_14_3_3'].iloc[i])
        dt.append(df['datetime'].iloc[i])
    if df['label2'].iloc[i]== 'e':
        ex.append(df['stochk_14_3_3'].iloc[i])
        et.append(df['datetime'].iloc[i])
    if df['label2'].iloc[i]== 'f':
        fx.append(df['stochk_14_3_3'].iloc[i])
        ft.append(df['datetime'].iloc[i])
        
Y=[ax,bx,cx,dx,ex,fx]
X=[at,bt,ct,dt,et,ft]
symbols=['circle','square','diamond','cross','triangle-up','triangle-down']
events=['%k touch 80','%d touch 80','%k touch 20','%d touch 20','cross-over in top20','cross-over in bottom20']
for i in range(0,4):
    fig.append_trace(
    go.Scatter(
        x=X[i],
        y=Y[i],
        marker_symbol=symbols[i],
        mode="markers",
        marker_size=5,
        name=events[i],
    ), row=2, col=1  #  <------------ lower chart
)
    

# Extend our y-axis a bit
fig.update_yaxes(range=[-10, 110], row=2, col=1)
# Add upper/lower bounds
fig.add_hline(y=0, col=1, row=2, line_color="#666", line_width=2)
fig.add_hline(y=100, col=1, row=2, line_color="#666", line_width=2)
# Add overbought/oversold
fig.add_hline(y=20, col=1, row=2, line_color='#336699', line_width=2, line_dash='dash')
fig.add_hline(y=80, col=1, row=2, line_color='#336699', line_width=2, line_dash='dash')
# Make it pretty
layout = go.Layout(
    plot_bgcolor='#efefef',
    # Font Families
    font_family='Monospace',
    font_color='#000000',
    font_size=10,
    xaxis=dict(
        rangeslider=dict(
            visible=False
        )
    )
)
fig.update_layout(layout)
# View our chart in the system default HTML viewer (Chrome, Firefox, etc.)
fig.show()

```
