```python
def stats_summary(data, ws):
    # show trades statistical information the order is base on the template file
    # return a pd.Series the record all the number (# of buy/sell/count) (do not change the order)
    # last row of code output an csv file, and use the output copy paste the template excel file

    # GRYP
    b = 'YPGR   PYGR	RGPY	RGYP	PYRG	GRYP	GYPR	YGPR	RPYG	YPRG	RPGY	GRPY	GPYR	RYPG	PGYR	GYRP	PRYG	RYGP	YRPG	GPRY	YGRP	PRGY	PGRY YRGP'
    cols = b.split()
    result_ls = []
    index = 0
    for col in cols:
        #print(data[(data['event'] == col) & (data['ws10_pt10_sl8'] =='buy')].shape[0]) 
        buy_times = data[(data['event'] == col) & (data[ws] =='buy')].shape[0] # numbers of buy
        result_ls.append(buy_times)

        sell_times = data[(data['event'] == col) & (data[ws] =='sell')].shape[0] # numbers of sell
        result_ls.append(sell_times)

        count = data[(data['event'] == col)].shape[0] # numbers of event
        result_ls.append(count)


    #ZigZag753
    z_cols = [1, -1]
    for z_col in z_cols:
        buy_times = data[(data['zigzag753'] == z_col) & (data[ws] =='buy')].shape[0] # numbers of buy
        result_ls.append(buy_times)

        sell_times = data[(data['zigzag753'] == z_col) & (data[ws] =='sell')].shape[0] # numbers of sell
        result_ls.append(sell_times)

        count = data[(data['zigzag753'] == z_col)].shape[0] # numbers of event
        result_ls.append(count)

    #ZigZag114103
    z_cols = [1, -1]
    for z_col in z_cols:
        buy_times = data[(data['zigzag114103'] == z_col) & (data[ws] =='buy')].shape[0] # numbers of buy
        result_ls.append(buy_times)

        sell_times = data[(data['zigzag114103'] == z_col) & (data[ws] =='sell')].shape[0] # numbers of sell
        result_ls.append(sell_times)

        count = data[(data['zigzag114103'] == z_col)].shape[0] # numbers of event
        result_ls.append(count)

    # bb event
    bb_cols = ['bb_event1','bb_event2',	'bb_event3','bb_event4','bb_event5','bb_event6','bb_event7','bb_event8']
    for b_col in bb_cols:
        buy_times = data[(data[b_col] == 1) & (data[ws] =='buy')].shape[0] # numbers of buy
        result_ls.append(buy_times)

        sell_times = data[(data[b_col] == 1) & (data[ws] =='sell')].shape[0] # numbers of sell
        result_ls.append(sell_times)

        count = data[(data[b_col] == 1)].shape[0] # numbers of event
        result_ls.append(count)

    #MACD
    MACD_ls = ['dir_change', 'two_peaks']
    for i in MACD_ls:
        if i == 'dir_change':
            buy_times = data[(data[i] == 1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(buy_times)

            sell_times = data[(data[i] == 1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(sell_times)

            count = data[(data[i] == 1)].shape[0] # numbers of event
            result_ls.append(count)

            buy_times = data[(data[i] == -1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(buy_times)

            sell_times = data[(data[i] == -1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(sell_times)

            count = data[(data[i] == -1)].shape[0] # numbers of event
            result_ls.append(count)

        if i == 'two_peaks':
            buy_times = data[(data[i] == 1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(buy_times)

            sell_times = data[(data[i] == 1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(sell_times)

            count = data[(data[i] == 1)].shape[0] # numbers of event
            result_ls.append(count)

    #SO1
    so_ls = ['SO1' ,"SO2"]
    for col in so_ls:
        if col=='SO1':
            for j in range(4):
                buy_times = data[(data[col] == j+1) & (data[ws] =='buy')].shape[0] # numbers of buy
                result_ls.append(buy_times)

                sell_times = data[(data[col] == j+1) & (data[ws] =='sell')].shape[0] # numbers of sell
                result_ls.append(sell_times)

                count = data[(data[col] == j+1)].shape[0] # numbers of event
                result_ls.append(count)
        
        if col=='SO2':
            for j in range(2):
                buy_times = data[(data[col] == j+1) & (data[ws] =='buy')].shape[0] # numbers of buy
                result_ls.append(buy_times)

                sell_times = data[(data[col] == j+1) & (data[ws] =='sell')].shape[0] # numbers of sell
                result_ls.append(sell_times)

                count = data[(data[col] == j+1)].shape[0] # numbers of event
                result_ls.append(count)

    #MA
    for j in range(1):
        buy_times = data[(data['label'] ==1) & (data[ws] =='buy')].shape[0] # numbers of buy
        result_ls.append(buy_times)

        sell_times = data[(data['label'] ==1) & (data[ws] =='sell')].shape[0] # numbers of sell
        result_ls.append(sell_times)

        count = data[(data['label'] == 1)].shape[0] # numbers of event
        result_ls.append(count)

    #PSAR
    psar_ls = ['Trend', "Change", 'Break']
    for col in psar_ls:
        if col == 'Trend':
            buy_times = data[(data[col] ==1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(buy_times)

            sell_times = data[(data[col] ==1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(sell_times)

            count = data[(data[col] == 1)].shape[0] # numbers of event
            result_ls.append(count)
        if col == 'Change':
            buy_times = data[(data[col] ==1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(buy_times)

            sell_times = data[(data[col] ==1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(sell_times)

            count = data[(data[col] == 1)].shape[0] # numbers of event
            result_ls.append(count)

            sbuy_times = data[(data[col] ==-1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(sbuy_times)

            ssell_times = data[(data[col] ==-1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(ssell_times)

            scount = data[(data[col] == -1)].shape[0] # numbers of event
            result_ls.append(scount)
        if col =='Break':
            buy_times = data[(data[col] ==1) & (data[ws] =='buy')].shape[0] # numbers of buy
            result_ls.append(buy_times)

            sell_times = data[(data[col] ==1) & (data[ws] =='sell')].shape[0] # numbers of sell
            result_ls.append(sell_times)

            count = data[(data[col] == 1)].shape[0] # numbers of event
            result_ls.append(count)

    #print(f'Result ls length: {len(result_ls)}')

    series = pd.Series(result_ls, name = ws)
    return series
```

```python
def trade_summary(data,window_size,pt,sl, event,event_value,one_pip):
    # calculate winning and losing trade stats return a pd.Series index base on the test_parameter

    parameter = f'ws{window_size}_pt{pt}_sl{sl}'
    data.fillna('',inplace = True)
    #data[parameter] = ''

    wining_trade = 0
    losing_trade = 0

    action_ls = []

    for row in tqdm(range(0,len(data) -1)):

        current_open,current_close,current_low,current_high = data['open'][row], data['close'][row], data['low'][row], data['high'][row]

        if data[event][row] == event_value:

                enter_price =  current_close
                max_given_ws = data['high'].iloc[row+1 : row+window_size+1].max()
                min_given_ws = data['low'].iloc[row+1: row+window_size+1].min()

                max_seq = data['high'].iloc[row+1 : row+window_size+1]
                min_seq =  data['low'].iloc[row+1: row+window_size+1]
                
                max_index = max_seq[max_seq==max_given_ws].index[0]
                min_index = min_seq[min_seq==min_given_ws].index[0]
                
                if (max_given_ws - enter_price)/one_pip >= pt and (enter_price - min_given_ws)/one_pip<sl:
                    wining_trade+=1

                elif (max_given_ws - enter_price)/one_pip >= pt and (enter_price - min_given_ws)/one_pip>=sl:
                    if max_index<= min_index:
                        wining_trade+=1
                    else:
                        losing_trade+=1

                elif (enter_price - min_given_ws)/one_pip>= pt and (max_given_ws - enter_price)/one_pip<sl:
                    wining_trade+=1
                
                elif (enter_price - min_given_ws)/one_pip>= pt and (max_given_ws - enter_price)/one_pip>=sl:
                    if min_index<= max_index:
                        wining_trade+=1
                    else:
                        losing_trade+=1

                elif ((max_given_ws - enter_price)/one_pip < pt) or (enter_price - min_given_ws)/one_pip < pt:
                    losing_trade+=1

    print(f'number of winning trade: {wining_trade} \nnumber of losing trade: {losing_trade} \nwin rate: {wining_trade/(wining_trade+losing_trade):.2f}')
    if losing_trade>0 or wining_trade>0:
        win_rate = round(wining_trade/(losing_trade+wining_trade), 2)
        dictionary = {parameter:[wining_trade, losing_trade, win_rate]}
    else:
        dictionary = {parameter:[wining_trade, losing_trade, 0]}
    return dictionary
    
# no window size
def trade_summary_nows(data,pt,sl,event,event_value,one_pip):
    # calculate winning and losing trade stats return a pd.Series index base on the test_parameter

    parameter = f'nows_pt{pt}_sl{sl}'
    data.fillna('',inplace = True)
    #data[parameter] = ''
    one_pip = 0.0001
    
    wining_trade = 0
    losing_trade = 0


    for row in tqdm(range(0,len(data) -1)):
        not_found = True
        current_open,current_close,current_low,current_high = data['open'][row], data['close'][row], data['low'][row], data['high'][row]

        if data[parameter][row] == '' and row<=len(data):
            index = 1
            enter_price =  current_close

            while not_found:
                max_given_ws = data['high'].iloc[row+1 : row+1+index].max()
                min_given_ws = data['low'].iloc[row+1 : row+1+index].min()

                max_seq = data['high'].iloc[row+1 : row+1+index]
                min_seq =  data['low'].iloc[row+1 : row+1+index]
                #print(max_seq)
                max_index = max_seq[max_seq==max_given_ws].index[0]
                min_index = min_seq[min_seq==min_given_ws].index[0]
                index+=1

                if (max_given_ws - enter_price)/one_pip >= pt and (enter_price - min_given_ws)/one_pip<sl:
                    wining_trade+=1
                    not_found = False

                elif (max_given_ws - enter_price)/one_pip >= pt and (enter_price - min_given_ws)/one_pip>=sl:
                    if min_index<=max_index:
                        wining_trade+=1
                        not_found = False
                    else:
                        losing_trade+=1
                        not_found = False

                elif (enter_price - min_given_ws)/one_pip>= pt and (max_given_ws - enter_price)/one_pip<sl:
                    wining_trade+=1
                    not_found = False

                elif (enter_price - min_given_ws)/one_pip>= pt and (max_given_ws - enter_price)/one_pip>=sl:
                    if max_index<=min_index:
                        wining_trade+=1
                        not_found = False
                    else:
                        losing_trade+=1
                        not_found = False
                not_found = False
    print(f'number of winning trade: {wining_trade} \nnumber of losing trade: {losing_trade} \nwin rate: {wining_trade/(wining_trade+losing_trade):.2f}')
    if losing_trade>0 or wining_trade>0:
        win_rate = round(wining_trade/(losing_trade+wining_trade), 2)
        dictionary = {parameter:[wining_trade, losing_trade, win_rate]}
    else:
        dictionary = {parameter:[wining_trade, losing_trade, 0]}
    return dictionary
    
    
```
